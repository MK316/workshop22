## [2] Preprocessing & Word Frequency: 
- removing stop words, 
- selecting words with more than 3 characters

```python
#@markdown üìï Preprocessing Remove stopwords, numbers, < 2 character words
import re
text = text.lower()
retokenize =RegexpTokenizer("[\w]+")
words = retokenize.tokenize(text)

numbers = re.findall('[0-9]+[0-9]+',text)
words = [w for w in words if not w in numbers]
words = [w for w in words if len(w) > 2]

#@markdown - Tokenized words after filtering stop words & words with more than 3 letters:
print('‚õÑ Number of words before stopwords: %d'%len(words))
words = [w for w in words if not w in new_stopwords_list]
print('‚õÑ Number of words after stopwords: %d'%len(words))
```

```python
#@markdown Save the frequency total as 'freqlist_total.csv'
#@markdown - data = df0
#@markdown - Show final 5 items.
import pandas as pd

fd = nltk.FreqDist(words).most_common()
df0 = pd.DataFrame()
wlist = []
freq = []
n = len(fd)
for i in range(0,n):
  wlist.append(fd[i][0])
  freq.append(fd[i][1])

df0['ID'] = range(1,n+1)
df0['Words'] = wlist
df0['Freq'] = freq

df0.to_csv(r'/content/freqlist_total.csv',index=False)

df0.tail()
```

```python
#@markdown üö©üö© Warning: Run this code only once (or delete txtdata, tagged_txt folders before rerun this code.)

#@markdown Step1: Making a new folder named "txtdata" for corpus toolkit

os.mkdir("txtdata")

#@markdown Step 2: Changing directory to "txtdata" and save a file there.
%cd /content/txtdata/

#@markdown Step 3: Read a file from URL
url = "https://raw.githubusercontent.com/MK316/Class_Spring2022/main/Ch01.Visual.Village.txt"
os.system("curl " + url + " > txtdata.txt")

#@markdown Step 4: Save the file under the folder 'txtdata' as 'txtdata.txt'
file = open("txtdata.txt")
text = file.read().replace("\n", " ")
file.close()

#@markdown Step 5: Changing directory to /content
%cd /content/

#@markdown Step 6: Tokenize
mydata = ct.ldcorpus("txtdata") #load and read text files under 'txtdata' directory
tok_corp = ct.tokenize(mydata) #tokenize corpus - by default this lemmatizes as well

#@markdown Step 7: Getting frequency (with tag)
mydata = ct.frequency(mydata) #creates a frequency dictionary

#@markdown Step 8: Writing tagged data as '1.txt' under 'tagged_txt' folder
ct.write_corpus("tagged_txt",ct.tag(ct.ldcorpus("txtdata")))

tagged_freq = ct.frequency(ct.reload("tagged_txt"))
print("Top 10 tagged words: %s"%ct.head(tagged_freq, hits = 10))

#@markdown Step 9: Creating a dataframe (csv format) as a file: df4
import pandas as pd

data_dict = tagged_freq
data_items = data_dict.items()
data_list = list(data_items)
df4 = pd.DataFrame(data_list)
df4.columns = ["Tagged","Freq"]

tagged = df4["Tagged"]
pos = []
word = []

for i in range(0, len(tagged)):
  w = tagged[i]
  ws = w.split("_")
  word.append(ws[0])
  pos.append(ws[1])

# Add new columns to the dataframe

df4["POS"] = pos
df4["Word"] = word

# Rearranging column order (remove Tagged column)
cols = ["POS","Word","Freq"]
df4 = df4[cols]

#@markdown Step 10: Sort by POS and Freq
df4 = df4.sort_values(by=['POS', 'Freq'], ascending = False)


#@markdown Step 11: Save the dataframe as csv file under the folder "/content"
#@markdown File name: tagged_mytext.csv
df4.to_csv('tagged_mytext.csv')

df4.head()
```
